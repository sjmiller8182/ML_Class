```python
from agent import Agent
from monitor import interact
import gym
import numpy as np
import matplotlib.pyplot as plt

trials = 50000
gammas = [0.1,0.3,0.5,0.7,0.9]
alphas = [0.0025,0.005,0.0075,0.01,0.0125,0.015,0.0175]

avg_r = dict()
best_avg_r = dict()
results = [avg_r, best_avg_r]

for g in gammas:
    env = gym.make('Taxi-v2')
    agent = Agent(gamma = g)
    avg_rewards, best_avg_reward = interact(env, agent, num_episodes=trials)
    results[0][str(g)] = avg_rewards
    results[1][str(g)] = best_avg_reward
```

```python
d = results[0]
for key in d.keys():
    td = d[key]
    plt.plot(np.arange(len(td)),td, label = 'key');
```

```python
ind = np.arange(1, len(results[1].keys())+1)
labs = []
vals = []
best_g = ''
best_val = -500

for key in results[1].keys():
    labs.append(key)
    vals.append(results[1][key])
    if results[1][key] >= best_val:
        best_val = results[1][key]
        best_g = key

fig, ax = plt.subplots();
plt.bar(ind, vals);
ax.set_xticks(ind);
ax.set_xticklabels(labs);
ax.set_xlabel('gamma');
ax.set_ylabel('Reward');
ax.set_title('Reward of Agent in Taxi-V2 over ' + str(trials) + ' trials');

print('The best gamma is ' + best_g + ' with a reward of ' + str(best_val))
```

```python
avg_r = dict()
best_avg_r = dict()
results = [avg_r, best_avg_r]

for a in alphas:
    env = gym.make('Taxi-v2')
    agent = Agent(gamma = np.float64(best_g), alpha = a)
    avg_rewards, best_avg_reward = interact(env, agent, num_episodes=trials)
    results[0][str(a)] = avg_rewards
    results[1][str(a)] = best_avg_reward
```

```python
ind = np.arange(1, len(results[1].keys())+1)
labs = []
vals = []
best_g = ''
best_val = -500

for key in results[1].keys():
    labs.append(key)
    vals.append(results[1][key])
    if results[1][key] >= best_val:
        best_val = results[1][key]
        best_g = key

fig, ax = plt.subplots();
plt.bar(ind, vals);
ax.set_xticks(ind);
ax.set_xticklabels(labs);
ax.set_xlabel('alpha');
ax.set_ylabel('Reward');
ax.set_title('Reward of Agent in Taxi-V2 over ' + str(trials) + ' trials');

print('The best gamma is ' + best_g + ' with a reward of ' + str(best_val))
```
